```python
import pandas as pd


# Podemos acceder a los elementos (columnas) de un dataframe haciendo: dataframe.columnName o dataframe['columnName']
## dataframa[columnname][index] para acceder a un registro
```

```python
trucks = pd.read_csv('truck_event_text_partition.csv', index_col = 0)
```

```python
trucks.head()
```

```python
## Para seleccionar la primera fila del dataframe

trucks.iloc[0]
```

```python
trucks.iloc[1]
```

```python
### Para obtener la columna hacemos:

trucks.iloc[:, 0]
```

```python
## Si quiero obtener solo los diez primeros registros

trucks.iloc[:10, 0]
```

```python
### O puedo hacer:

trucks.iloc[0:10, 0]
```

```python
### Tamvién es posible pasar una lista

trucks.iloc[[1, 4, 5], 0]
```

```python
### los últimos registros...

trucks.iloc[-10:, [0, 1, 2]]
```

```python
trucks.iloc[-10:]
```

```python
trucks.columns
```

```python
## Ahora probamos LOC

trucks.loc[0, 'driverId']
```

```python
### Para seleccionar varias columnas


trucks.loc[:, ['driverId', 'eventTime', 'driverName']]
```

```python
'''iloc uses the Python stdlib indexing scheme, where the first element
of the range is included and the last one excluded. So 0:10 will select 
entries 0,...,9. loc, meanwhile, indexes inclusively. So 0:10 will select entries 0,...,10.'''
```

```python
### MANIPULATING THE INDEX ###
```

```python
### podemos usar el método set_index() para establecer un índice

trucks.set_index('driverName')
```

```python
### CONDITIONAL SELECTION ####
```

```python
trucks.eventType == 'Normal'
```

```python
## Podemos usar el método LOC con los condicionales:

trucks.loc[trucks.truckId > 50]
```

```python
## Podemos usar más de un condicional:

trucks.loc[(trucks.truckId > 50) & (trucks.driverId < 25)] ## Usamos | para el "or"
```

```python
''' isin. isin is lets you select data whose value "is in" a list of values.'''
```

```python
## Para seleccionar los datos de solo dos conductores...

trucks.loc[trucks.driverName.isin(['Jamie Engesser', 'Nadeem Asghar'])]
```

```python
'''The second is isnull (and its companion notnull). These methods let you highlight values which are (or are not) 
empty (NaN). For example, to filter out wines lacking a price tag in the dataset, here's what we would do:'''
```

```python
trucks.loc[trucks.driverId.notnull()]
```
