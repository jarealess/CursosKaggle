```python
import pandas as pd

trucks = pd.read_csv('truck_event_text_partition.csv', index_col = 0)
```

```python
trucks.columns
```

```python
trucks.driverName.value_counts()
```

```python
trucks.groupby('driverName').driverName.count()
```

```python
### ES LO MISMO QUE HACER ###

trucks.groupby('driverName').size()
```

```python
trucks.groupby('driverName').count()
```

```python
### Podemos hacer otras cosas como mirar el mínimo/máximo ID del cminón para cada conductor

trucks.groupby('driverName').truckId.max()
```

```python
## Se puede agrupar por más de una categoría


trucks.groupby(['driverName', 'routeName']).truckId.count()
```

```python
## Podemos usar el método --> AGG <-- para aplicar varias operaciones con grupby


trucks.groupby('driverName').truckId.agg([min, max, len])
```

```python
## SORTING ###

ordenada = trucks.groupby('driverName').truckId.agg([min, max, len])
ordenada.sort_values(by='len')
```

```python
ordenada.sort_values(by='len', ascending = False)
```

```python
### SORT BY INDEX

ordenada.sort_index()
```

```python
### Se puede ordenar por más de una columna

ordenada.sort_values(by=['min', 'len'])
```
