```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
insurance_data = pd.read_csv('insurance.csv')
```

```python
print(insurance_data.shape)
insurance_data.head()
```

```python
## creating a scatter plot

sns.scatterplot(x=insurance_data.bmi, y=insurance_data.charges)
```

```python
## REGRESSION PLOT ###
sns.regplot(x=insurance_data['bmi'], y=insurance_data['charges'])
```

BMI and charges a possitively correlated

```python
## Color the poins by a column
sns.scatterplot(x=insurance_data['bmi'], y=insurance_data['charges'], hue=insurance_data.smoker)
```

```python
## Si queremos aadir una regresin por cada categora
sns.lmplot(x="bmi", y="charges", hue="smoker", data=insurance_data)
```

```python
## Color the poins by a column
sns.scatterplot(x=insurance_data['bmi'], y=insurance_data['charges'], hue=insurance_data.sex)
```

```python
## Probemos si se puede de otra forma la doble regresin
sns.lmplot(x=insurance_data.bmi, y=insurance_data.charges, hue=insurance_data.smoker)
```

```python
## Color the poins by a column
sns.scatterplot(x=insurance_data['bmi'], y=insurance_data['charges'], hue=insurance_data.region)
```
